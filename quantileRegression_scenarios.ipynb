```python
import numpy as np
from functions import plot_and_compare_errors
```

# Scenario 1: Outliers

```python
np.random.seed(0)
X_outliers = np.random.uniform(0, 10, 100)
y_outliers = 2 * X_outliers + np.random.normal(0, 1, 100)
y_outliers[95:] += 15  # Add outliers
plot_and_compare_errors(X_outliers, y_outliers, quantiles=[0.25, 0.5, 0.75], 
                        scenario_title="Scenario 1: Outliers")
```

# Scenario 2: Heteroscedasticity

```python
np.random.seed(1)
X_hetero = np.random.uniform(0, 10, 100)
y_hetero = 2 * X_hetero + np.random.normal(0, X_hetero)  # Increasing variance with X
plot_and_compare_errors(X_hetero, y_hetero, quantiles=[0.25, 0.5, 0.75], 
                        scenario_title="Scenario 2: Heteroscedasticity")
```

# Scenario 3: Non-linear Relationship

```python
np.random.seed(2)
X_nonlinear = np.random.uniform(0, 10, 100)
y_nonlinear = 2 * np.sqrt(X_nonlinear) + np.random.normal(0, 0.5, 100)
plot_and_compare_errors(X_nonlinear, y_nonlinear, quantiles=[0.25, 0.5, 0.75], 
                        scenario_title="Scenario 3: Non-linear Relationship")
```

# Scenario 4: Non-uniform Residuals

```python
np.random.seed(3)
X_nonuniform = np.random.uniform(0, 10, 100)
y_nonuniform = 2 * X_nonuniform + np.random.normal(0, 0.5, 100)
y_nonuniform[:50] -= 2  # Shifted residuals for half the data
plot_and_compare_errors(
    X_nonuniform,
    y_nonuniform,
    quantiles=[0.25, 0.5, 0.75],
    scenario_title="Scenario 4: Non-uniform Residuals",
)
```

# Scenario 5: Salaries with Non-Uniform Growth

```python
np.random.seed(4)
X_salaries = np.random.uniform(0, 20, 200)  # Years of experience
# Non-linear and asymmetric salary growth
y_salaries = 10 + 2 * X_salaries + 0.1 * X_salaries**2 + np.random.normal(0, 5, 200)
y_salaries[X_salaries > 15] += 50  # Boost for high experience (executives)
plot_and_compare_errors(
    X_salaries,
    y_salaries,
    quantiles=[0.25, 0.5, 0.75],
    scenario_title="Scenario 5: Salaries with Non-Uniform Growth",
)
```
